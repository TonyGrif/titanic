```python
# Initialize notebook
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../resources/train.csv")
```

```python
# Class Survival Rate
df[["Pclass", "Survived"]] \
    .groupby("Pclass") \
    .value_counts()
```

```python
# Sex Survival Rate
df[["Survived", "Sex"]] \
    .groupby("Sex") \
    .value_counts()
```

```python
# Age Survival Rate
bins = [x for x in range(0, 100, 10)]

df["Survived"].groupby(pd.cut(df["Age"], bins=bins), observed=True) \
    .value_counts()
```

```python
# Fare Survival Rate
df[["Survived", "Fare"]] \
    .groupby("Fare") \
    .value_counts()
```

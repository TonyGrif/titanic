```python
# Initialize notebook
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../resources/train.csv")
```

```python
# Class Survival Rate
ax = df[["Pclass", "Survived"]] \
    .groupby("Pclass") \
    .value_counts() \
    .unstack() \
    .plot(kind="bar", stacked=True, title="Class Survival Breakdown", xlabel="Class Designation")
```

```python
# Sex Survival Rate
ax = df[["Survived", "Sex"]] \
    .groupby("Sex") \
    .value_counts() \
    .unstack() \
    .plot(kind="bar", stacked=True, title="Sex Survival Breakdown", xlabel="Sex of Passengers")
```

```python
# Age Survival Rate
bins = [x for x in range(0, 100, 10)]

ax = df["Survived"] \
    .groupby(pd.cut(df["Age"], bins=bins), observed=True) \
    .value_counts() \
    .unstack() \
    .plot(kind="bar", stacked=True, title="Age Suvival Breakdown", xlabel="Age Bin")
```

```python
# Fare Survival Rate
df[["Survived", "Fare"]] \
    .groupby("Fare") \
    .value_counts()
```

```python
# Embarked Suvival Rate
ax = df[["Survived", "Embarked"]] \
    .groupby("Embarked") \
    .value_counts() \
    .unstack() \
    .plot(kind="bar", stacked=True, title="Embarking Port Surival Ratio", xlabel="Port Designation")
```

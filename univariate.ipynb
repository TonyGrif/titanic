```python
import matplotlib.pyplot as plt
import pandas

df = pandas.read_csv("train.csv")
```

```python
# Died vs Survived Ratio

survivalDf = df["Survived"].replace({0: "Died", 1: "Surived"}).value_counts()

labels = ("Died", "Survived")
fig, ax = plt.subplots()
ax.pie(survivalDf, labels=labels, autopct='%1.1f%%')
```

```python
# Ticket Class Ratio

classDf = df["Pclass"] \
  .replace({1: "First Class", 2: "Second Class", 3: "Third Class"}) \
  .value_counts()

fig, ax = plt.subplots()
ax.pie(classDf, labels=classDf.keys(), autopct='%1.1f%%')
```

```python
# Sex Ratio

sexDf = df["Sex"].value_counts()

fig, ax = plt.subplots()
ax.pie(sexDf, labels=sexDf.keys(), autopct='%1.1f%%')
```

```python
# Age Makeup

print(f"Null count on age: {df['Age'].isna().sum()}")

ageDf = df["Age"].dropna().value_counts().sort_index()
plt.plot(ageDf.keys(), ageDf.values.tolist())
```

```python
# Sibling and Spouse Numbers

sibDf = df["SibSp"].value_counts()
plt.bar(sibDf.keys(), sibDf.values.tolist())
```

```python
# Parent and Children Numbers

sibDf = df["Parch"].value_counts()
plt.bar(sibDf.keys(), sibDf.values.tolist())
```

```python
# Fare Analysis

fareDF = df["Fare"].value_counts().sort_index()
plt.plot(fareDF.keys(), fareDF.values.tolist())
```

```python
# Embarked Ratio

embarkedDf = df["Embarked"] \
  .replace({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"}) \
  .value_counts()

fig, ax = plt.subplots()
ax.pie(embarkedDf, labels=embarkedDf.keys(), autopct='%1.1f%%')
```

```python
# Initialize notebook
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../resources/train.csv")
```

```python
# Utility functions
def pie(title, values, labels) -> None:
    fig, ax = plt.subplots()
    plt.title(title)
    ax.pie(values, labels=labels, autopct="%1.1f%%")

def bar(title, xcord, height) -> None:
    fix, ax = plt.subplots()
    plt.title(title)
    ax.bar(xcord, height)
```

```python
# Survival Ratio
survival = df["Survived"] \
    .value_counts() \
    .rename({0: "Died", 1: "Survived"})

pie("Survival Ratio", survival.values, survival.keys())
```

```python
# Socio-Economic Status Ratio
socec = df["Pclass"] \
    .value_counts() \
    .rename({1: "Upper", 2: "Middle", 3: "Lower"})

pie("Socio-Economic Breakdown", socec.values, socec.keys())
```

```python
# Sex Ratio
sex = df["Sex"].value_counts()

pie("Sex Breakdown", sex.values, sex.keys())
```

```python
# Age Graph
bins = [x for x in range(0, 100, 10)]
age = df.groupby(pd.cut(df["Age"], bins=bins), observed=True).size()

bar("Age Demographic Count", age.keys().astype(str), age.values)
```

```python
# Traveling with Sibling / Spouse Ratio
sibsp = df["SibSp"] \
    .value_counts() \
    .sort_index()

bar("Count of Passengers Traveling with x Siblings/Spouses", sibsp.keys().astype(str), sibsp.values)
```

```python
# Traveling with Parent / Child Ratio
parch = df["Parch"] \
    .value_counts() \
    .sort_index()

bar("Count of Passengers Traveling with x Parents/Children", parch.keys().astype(str), parch.values)
```

```python
# Ticket Fare Graph
fare = df["Fare"] \
    .round(0) \
    .astype(int) \
    .value_counts()

bar("Ticket Prices", fare.values, fare.keys())
```

```python
# Embarked Ratio
embarked = df["Embarked"] \
    .value_counts() \
    .rename({"C": "Cherbourg", "Q": "Queenstown", "S": "Southhampton"})

pie("Embarking Location", embarked.values, embarked.keys())
```

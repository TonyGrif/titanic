```python
# Initialize notebook

import pandas as pd

df = pd.read_csv("../resources/train.csv")
```

```python
# Survival Ratio
df["Survived"] \
    .value_counts() \
    .rename({0: "Died", 1: "Survived"})
```

```python
# Socio-Economic Status Ratio
df["Pclass"] \
    .value_counts() \
    .rename({1: "Upper", 2: "Middle", 3: "Lower"})
```

```python
# Sex Ratio
df["Sex"].value_counts()
```

```python
# Age Graph
bins = [x for x in range(0, 100, 10)]
df.groupby(pd.cut(df["Age"], bins=bins), observed=True).size()
```

```python
# Traveling with Sibling / Spouse Ratio
df["SibSp"] \
    .value_counts() \
    .sort_index()
```

```python
# Traveling with Parent / Child Ratio
df["Parch"] \
    .value_counts() \
    .sort_index()
```

```python
# Ticket Fare Graph
df["Fare"] \
    .round(0) \
    .astype(int) \
    .value_counts()
```

```python
# Embarked Ratio
df["Embarked"] \
    .value_counts() \
    .rename({"C": "Cherbourg", "Q": "Queenstown", "S": "Southhampton"})
```
